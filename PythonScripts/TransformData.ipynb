```python
lines = []
lineNumber = 0
with open("C:\\Projects\\TrainDatasetGenerator\\2021-05-05_09-26-20.csv") as fin:
    for line in fin:
        split_line = line.split(sep=';' ,maxsplit=450)
        result_line = []
        for i in range(0, 450, 2):
            if(split_line[i] == '0' and split_line[i+1] == '0'):
                result_line.append('0')
            elif(split_line[i] == '1' and split_line[i+1] == '0'):
                result_line.append('1')
            elif(split_line[i] == '0' and split_line[i+1] == '1'):
                result_line.append('-1')
            else:
                raise Exception(f"Something went wrong: \n line {split_line} \n index {i}")
        result_line.append(split_line[450])
        lines.append(';'.join(result_line))
        print("line number: ", lineNumber)
        lineNumber += 1
```

```python
with open("C:\\Projects\\TrainDatasetGenerator\\v2_2021-05-05_09-26-20.csv", "w+") as fout:
    fout.writelines(lines)
```
